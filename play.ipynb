```python
import rushlite as rsl
```

```python
N = 100
D = 10

x = rsl.rand([N, D], requires_grad=False)
y = rsl.rand([N, 1], requires_grad=False)


w = rsl.rand([D, 1], requires_grad=True)
b = rsl.rand([1, 1], requires_grad=True)

for _ in range(10000):
    pred = x @ w + b
    loss = ((y - pred)**2).sum(axis=0).sum(axis=1)
    
    w.zero_grad()
    b.zero_grad()
    loss.backward()

    print(loss)

    w -= 0.001 * rsl.Variable(w.grad)
    b -= 0.001 * rsl.Variable(b.grad)

"""
friction log: 

need to cast w.grad to a variable first
> out: b.data -= 0.001 * rsl.Variable(b.grad)
> exp: b -= 0.001 * b.grad

sum() doesn't sum over all axes
> out: loss = ((y - pred)**2).sum(axis=0).sum(axis=1)
> exp: loss = ((y - pred)**2).sum()

"""
```
